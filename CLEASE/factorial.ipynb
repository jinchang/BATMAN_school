```python
import math 
  
print ("The factorial of 23 is : ", end="") 
print (math.factorial(23)) 
```

```python
def combinations(val):
    if val%2 == 1:
        return 0
    half = val/2
    return int(math.factorial(val)/(math.factorial(half)*math.factorial(half)))
    # return int(math.factorial(val))
def exponent(val):
    return 2**val
```

```python
print(combinations(4))
```

```python
size = []
comb = []
expo = []
for x in range(2,100,2):
    size.append(x)
    comb.append(combinations(x))
    expo.append(exponent(x))

print(size, comb)
```

```python
%matplotlib notebook
from matplotlib.figure import Figure
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
# ax.set_title(title)
ax.set_xlabel("No. atoms in cell")
ax.set_ylabel("No. combinations")
ax.semilogy(size, comb, 'bo', mfc='none')
# ax.semilogy(size, expo, 'ro', mfc='none')
```

```python
def combination36(val):
    if val > 36:
        return 0
    remain = 36 - val
    return int(math.factorial(36)/(math.factorial(val)*math.factorial(remain)))
    # return int(math.factorial(val))
```

```python
size = []
comb = []
for x in range(0,37):
    size.append(x)
    comb.append(combination36(x))

print(size, comb)
```

```python
%matplotlib notebook
from matplotlib.figure import Figure
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
# ax.set_title(title)
ax.set_xlabel("No. Cu atoms in 36 atom cell")
ax.set_ylabel("No. combinations")
ax.semilogy(size, comb, 'bo', mfc='none')
# ax.semilogy(size, expo, 'ro', mfc='none')
```
